In this Notebook, we'll connect to Azure ML Workspace and persist the configuration.

First, we need to make sure we installed and attached the **Azure Machine Learning SDK** package to our Databricks cluster.

See [this article](https://docs.microsoft.com/en-us/azure/machine-learning/service/how-to-configure-environment#azure-databricks) for instruction to attach the SDK as a library.

```
import azureml.core

# Check core SDK version number - based on build number of preview/master.
print("SDK version:", azureml.core.VERSION)
```

```
subscription_id = "<Your SubscriptionId>" #you should be owner or contributor
resource_group = "<Resource group - new or existing>" #you should be owner or contributor
workspace_name = "<workspace to be created>" #your workspace name
```

```
from azureml.core import Workspace

workspace = Workspace.get(workspace_name, subscription_id=subscription_id, resource_group=resource_group)

workspace.get_details()
```

```
%sh mkdir /dbfs/aml_ibm_config
```

```
# persist the subscription id, resource group name, and workspace name in aml_config/config.json.
workspace.write_config(path="/dbfs/aml_ibm_config")
```

```
# import the Workspace class and check the azureml SDK version
from azureml.core import Workspace

workspace = Workspace.from_config(path="/dbfs/aml_config/config.json")

print('Workspace name: ' + workspace.name, 
      'Azure region: ' + workspace.location, 
      'Subscription id: ' + workspace.subscription_id, 
      'Resource group: ' + workspace.resource_group, sep = '\n')
```
